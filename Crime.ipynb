```python
#import the modules
import pandas as pd
import os
from google.colab import drive
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/Data/'
lookup = pd.read_csv('/content/drive/MyDrive/lookup_msoa_lsoa.csv')
file_list = os.listdir(file_path)

all_dfs = []
for file in file_list:
  # Only process files, skip directories like .ipynb_checkpoints
  if os.path.isfile(os.path.join(file_path, file)):
    df_temp = pd.read_csv(os.path.join(file_path, file))
    all_dfs.append(df_temp)

df_append = pd.concat(all_dfs, ignore_index=True)

# Merge df_append (which contains crime data) with lookup (which maps LSOA to MSOA)
# to add the 'msoa' column to each crime record.
crime_lsoa_msoa = df_append.merge(lookup, left_on = 'LSOA name', right_on = 'LSOA21NM', how='left')

# Group by 'msoa' and count the number of crimes in each MSOA.
# Assuming the original ('lsoa','sum') was intended to count crimes per MSOA.
crime_msoa = crime_lsoa_msoa.groupby('MSOA21NM').agg(crime_count = ('LSOA name','count'))

crime_msoa.to_csv('crime_msoa.csv')
```
